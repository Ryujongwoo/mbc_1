```python
# stackModule 전체를 import 하지 않고 stackModule의 Stack 클래스만 import 시켜서 사용한다.
from stackModule import Stack
# Stack 클래스만 import 시켰으므로 위의 경우처럼 모듈 이름이나 별병을 붙일 필요없이 바로 사용하면 된다.
stack = Stack()
stack.view()
```

```python
menu = 0
while menu != 4:
    while True:
        print('=' * 31)
        print(' 1.push  2.pop  3.view  4.종료 ')
        print('=' * 31)
        menu = int(input('원하는 메뉴를 입력하고 엔터키를 누르세요: '))
        if menu in [1, 2, 3, 4]:
            break
        else:
            print('메뉴는 1 ~ 4 사이의 값만 입력해야 합니다.')
        # ===== if
    # ===== while True
    
    if menu == 1:
        data = input('스택에 저장할 데이터를 입력하세요: ')
        stack.push(data)
    elif menu == 2:
        stack.pop()
    elif menu == 3:
        stack.view()
# ===== while memu != 4
print('프로그램을 종료합니다. 바이바이~~~~~')
```
