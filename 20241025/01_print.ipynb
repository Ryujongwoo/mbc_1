'%'와 서식 문자를 서식을 지정해서 출력하기

서식 문자  
d(정수), f(실수), s(문자열)를 사용해서 출력 서식을 지정할 수 있고 출력 서식 문자를 제외한 나머지 문자는 입력한 그대로 출력된다.

출력 서식의 형식  
%[-][0][n][.m]서식문자 => [] 안의 내용은 필요에 따라 선택해서 사용한다.  
n: 전체 자리수, 전체 자리수가 지정되면 데이터는 오른쪽을 맞춰서 출력된다.  
-: 전체 자리수가 지정된 경우 데이터를 왼쪽으로 맞춰서 출력한다.  
0: 전체 자리수가 지정된 경우 숫자의 남는 자리를 '0'을 채워 출력한다. => 문자앨 데이터에서는 무시된다.  
.m: 실수를 출력하려는 경우 소수점 아래 자리수를 지정한다.

print('출력 서식' % (출력할 데이터, ...)) => 출력할 데이터가 2개 이상일 경우

```python
print('%d + %d = %d' % (5, 3, 5 + 3)) # 덧셈
print('%d - %d = %d' % (5, 3, 5 - 3)) # 뺄셈
print('%d * %d = %d' % (5, 3, 5 * 3)) # 곱셈
print(5 / 3) # 나눗셈, 파이썬은 '/' 연산자를 사용하면 소수점 아래까지 계산된다. 실수 나눗셈
print('%d / %d = %d' % (5, 3, 5 / 3))
print('%d / %d = %f' % (5, 3, 5 / 3))
print(5 // 3) # 나눗셈, 파이썬은 '//' 연산자를 사용하면 몫을 계산한다. 정수 나눗셈
print('%d // %d = %d' % (5, 3, 5 // 3))
print('%d // %d = %f' % (5, 3, 5 // 3))
print(5 % 3) # 나머지
print('%d %% %d = %d' % (5, 3, 5 % 3)) # 출력 서식에 '%'문자 자체를 출력하려면 '%%'와 같이 사용해야 한다.
print('%d ** %d = %d' % (5, 3, 5 ** 3)) # 거듭제곱
print('루트 %d = %f' % (2, 2 ** 0.5)) # 제곱근
```

format() 메소드와 출력할 데이터의 인덱스를 사용해 출력하기 => 인덱스는 0부터 시작한다.  
format() 메소드를 사용하면 데이터가 출력될 위치에 서식 문자를 사용하지 않고 {}를 사용한다.

```python
print('{} {}'.format('홍길동', '바보'))
# 인덱스를 지정하지 않은 상태에서 아래와 같이 {}가 출력할 데이터보다 많으면 에러가 발생된다.
# print('{} {} {}'.format('홍길동', '바보')) # 에러
# {}가 모자라면 에러는 발생되지 않고 출력되며 남는 데이터는 무시된다.
print('{}'.format('홍길동', '바보'))
```

```python
print('{0} {1}'.format('홍길동', '바보'))
print('{1} {0}'.format('홍길동', '바보'))
# {} 안에 인덱스를 지정하지 않은 상태에서 {}가 남으면 에러가 발생되었지만 {} 안에 인덱스를 지정하면
# {}와 데이터의 개수가 달라도 인덱스 범위만 벗나지 않는다면 정상 출력된다.
print('{0} {1} {0}'.format('홍길동', '바보'))
# {}에 지정한 인덱스가 인덱스의 범위를 벗어나면 에러가 발생된다.
# print('{0} {1} {2}'.format('홍길동', '바보')) # 에러
# {} 내부에 인덱스를 부분적으로 지정하면 에러가 발생된다.
# print('{0} {}'.format('홍길동', '바보')) # 에러
```

format() 메소드로 {인덱스:출력서식}을 사용하면 서식을 지정해서 출력할 수 있다.  
인덱스는 생략하고 출력 서식만 지정할 수 있는데 이 때, ':'을 생략하면 안되고 출력 서식에 '%'를 사용하지 않는다.

```python
print('%-10s는(은) %7d원 입니다.' % ('멜론', 15000))
print('{}는(은) {}원 입니다.'.format('멜론', 15000))
print('{0}는(은) {1}원 입니다.'.format('멜론', 15000))
print('{0:10s}는(은) {1:7d}원 입니다.'.format('멜론', 15000))
print('{:10s}는(은) {:7d}원 입니다.'.format('멜론', 15000))
print('{:10s}는(은) {:7,d}원 입니다.'.format('멜론', 15000))
print('{:10s}는(은) {:.2f}원 입니다.'.format('멜론', 15000))
print('{:10s}는(은) {:.2%}원 입니다.'.format('멜론', 15000))
print('{:10s}는(은) {:,.2%}원 입니다.'.format('멜론', 15000))
```

```python
# '<', '^', '>'를 사용해서 문자열 정렬 방식을 지정할 수 있다. => '<'가 기본값으로 사용된다.
print('{0:10s}는(은) {1:7d}원 입니다.'.format('멜론', 15000))
print('{0:<10s}는(은) {1:7d}원 입니다.'.format('멜론', 15000))
print('{0:^10s}는(은) {1:7d}원 입니다.'.format('멜론', 15000))
print('{0:>10s}는(은) {1:7d}원 입니다.'.format('멜론', 15000))
```

```python
# 문자열 출력시 남는 자리의 공백을 채울 채움 문자를 '<', '^', '>' 앞에 지정할 수 있다.
print('{0:_<10s}는(은) {1:7d}원 입니다.'.format('멜론', 15000))
print('{0:_^10s}는(은) {1:7d}원 입니다.'.format('멜론', 15000))
print('{0:_>10s}는(은) {1:7d}원 입니다.'.format('멜론', 15000))
```
