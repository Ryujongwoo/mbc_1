```python
# 이진 트리 클래스
class BinaryTree:
    def __init__(self, data=None):
        self.left = None # 왼쪽 자식 노드의 주소를 기억한다.
        self.data = data # 데이터
        self.right = None # 오른쪽 자식 노드의 주소를 기억한다.
    # ===== __init__()
    
    # 이진 트리에 데이터를 추가하는 함수
    def append(self, data):
        print(f'이진 트리에 추가하려는 데이터 {data}의 부모 노드는 {self.data}입니다.')
        
        # 이진 트리에 추가할 데이터와 부모 노드의 데이터를 비교해서 이진 트리에 데이터가 추가되는 위치가 결정된다.
        # 이진 트리에 추가할 데이터가 부모 노드의 데이터보다 작으면 부모 노드의 왼쪽에 추가한다.
        # 이진 트리에 추가할 데이터가 부모 노드의 데이터보다 크면 부모 노드의 오른쪽에 추가한다.
        
        # 이진 트리에 추가하려는 데이터가 부모 노드의 데이터보다 작은가?
        if data < self.data:
            # 부모 노드의 왼쪽에 데이터를 추가한다.
            print(f'이진 트리에 추가할 데이터({data})가 부모 노드의 데이터({self.data})보다 작기 때문에 부모 노드의 왼쪽에 삽입한다.')
            # 부모 노드의 왼쪽 링크(self.left)가 비어있어야(None) 데이터를 추가할 수 있다.
            if self.left is None: # 부모 노드의 왼쪽 링크가 비어있는가?
                # 부모 노드의 왼쪽 링크가 비어있으므로 데이터를 추가한다.
                print(f'부모({self.data}) 왼쪽 링크가 비어있으므로 데이터 {data} 추가 가능')
                # 이진 트리에 새 노드를 추가해야 하므로 이진 트리에 추가할 데이터를 BinaryTree 클래스 객체를 생성해서 부모 노드의
                # 비어있는 왼쪽 링크에 BinaryTree 클래스 객체의 주소를 넣어준다.
                self.left = BinaryTree(data)
                print(f'부모({self.data}) 왼쪽에 데이터 {data} 추가 완료')
            else:
                # 부모 노드의 왼쪽 링크가 비어있지 않기 때문에 데이터를 추가할 수 없다.
                print(f'부모({self.data}) 왼쪽 링크가 비어있지 않기 때문에 데이터 {data} 추가 불가능')
                # 부모 노드의 왼쪽 링크가 비어있지 않기 때문에 부모 노드의 왼쪽 링크가 참조하는 노드에서 append() 메소드를
                # 실행해야 한다.
                print(f'부모({self.data}) 왼쪽에 {data}를(을) 추가할 수 없으므로 부모 왼쪽 노드 {self.left.data}에서 append() 메소드를 실행해야 한다.')
                self.left.append(data)
            # ===== if
            
        # 이진 트리에 추가하려는 데이터가 부모 노드의 데이터보다 큰가?
        elif data > self.data:
            # 부모 노드의 오른쪽에 데이터를 추가한다.
            print(f'이진 트리에 추가할 데이터({data})가 부모 노드의 데이터({self.data})보다 크기 때문에 부모 노드의 오른쪽에 삽입한다.')
            # 부모 노드의 오른쪽 링크(self.right)가 비어있어야(None) 데이터를 추가할 수 있다.
            if self.right is None: # 부모 노드의 오른쪽 링크가 비어있는가?
                # 부모 노드의 오른쪽 링크가 비어있으므로 데이터를 추가한다.
                print(f'부모({self.data}) 오른쪽 링크가 비어있으므로 데이터 {data} 추가 가능')
                # 이진 트리에 새 노드를 추가해야 하므로 이진 트리에 추가할 데이터를 BinaryTree 클래스 객체를 생성해서 부모 노드의
                # 비어있는 오른쪽 링크에 BinaryTree 클래스 객체의 주소를 넣어준다.
                self.right = BinaryTree(data)
                print(f'부모({self.data}) 오른쪽에 데이터 {data} 추가 완료')
            else:
                # 부모 노드의 오른쪽 링크가 비어있지 않기 때문에 데이터를 추가할 수 없다.
                print(f'부모({self.data}) 오른쪽 링크가 비어있지 않기 때문에 데이터 {data} 추가 불가능')
                # 부모 노드의 오른쪽 링크가 비어있지 않기 때문에 부모 노드의 오른쪽 링크가 참조하는 노드에서 append() 메소드를
                # 실행해야 한다.
                print(f'부모({self.data}) 오른쪽에 {data}를(을) 추가할 수 없으므로 부모 오른쪽 노드 {self.right.data}에서 append() 메소드를 실행해야 한다.')
                self.right.append(data)
            # ===== if
            
        # 이진 트리에 추가하려는 데이터가 이진 트리에 존재하는가?
        else:
            # 이진 트리에 추가하려는 데이터가 이진 트리에 이미 존재하기 때문에 중복 메시지를 출력한다.
            print(f'{data}는(은) 이진 트리에 존재하는 데이터 입니다.')
        # ===== if
    # ===== append()
    
    # 이진 트리를 preorder 방식으로 운행(탐색)하는 함수, 부모 => 왼쪽 자식 => 오른쪽 자식
    def preorder(self):
        print('========== {:2d}의 preorder() 함수 시작 =========='.format(self.data))
        print(f'방문한 노드의 데이터: {self.data}')
        # 왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 preorder() 메소드를 실행한다.
        # if self.left != None:
        # if self.left is not None:
        # link(self.left, self.right)가 None이면 False, None이 아니면 True
        if self.left: # 왼쪽 자식이 있는가?
            self.left.preorder()
        # ===== if
        # 오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 preorder() 메소드를 실행한다.
        if self.right: # 오른쪽 자식이 있는가?
            self.right.preorder()
        # ===== if
        print('********** {:2d}의 preorder() 함수 종료 **********'.format(self.data))
    # ===== preorder()
    
    # 이진 트리를 inorder 방식으로 운행(탐색)하는 함수, 왼쪽 자식 => 부모 => 오른쪽 자식
    def inorder(self):
        print('========== {:2d}의 inorder() 함수 시작 =========='.format(self.data))
        # 왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 inorder() 메소드를 실행한다.
        if self.left: # 왼쪽 자식이 있는가?
            self.left.inorder()
        # ===== if
        print(f'방문한 노드의 데이터: {self.data}')
        # 오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 inorder() 메소드를 실행한다.
        if self.right: # 오른쪽 자식이 있는가?
            self.right.inorder()
        # ===== if
        print('********** {:2d}의 inorder() 함수 종료 **********'.format(self.data))
    # ===== inorder()
    
    # 이진 트리를 postorder 방식으로 운행(탐색)하는 함수, 왼쪽 자식 => 오른쪽 자식 => 부모
    def postorder(self):
        print('========== {:2d}의 postorder() 함수 시작 =========='.format(self.data))
        # 왼쪽 자식 노드가 있나 검사해서 있으면 왼쪽 자식 노드에서 postorder() 메소드를 실행한다.
        if self.left: # 왼쪽 자식이 있는가?
            self.left.postorder()
        # ===== if
        # 오른쪽 자식 노드가 있나 검사해서 있으면 오른쪽 자식 노드에서 postorder() 메소드를 실행한다.
        if self.right: # 오른쪽 자식이 있는가?
            self.right.postorder()
        # ===== if
        print(f'방문한 노드의 데이터: {self.data}')
        print('********** {:2d}의 postorder() 함수 종료 **********'.format(self.data))
    # ===== postorder()
```

```python
# 이진 트리의 최상위(root) 노드를 만든다.
root = BinaryTree(12)
# 이진 트리에 데이터를 추가한다.
root.append(12)
print('=' * 100)
root.append(6)
print('=' * 100)
root.append(20)
print('=' * 100)
root.append(3)
print('=' * 100)
root.append(10)
print('=' * 100)
root.append(15)
print('=' * 100)
root.append(25)
```

```python
# 이진 트리의 운행 => 같은 노드를 중복해서 방문하지 않고 모든 노드를 한 번씩 방문하는 방법
# 이진 트리의 운행은 부모 노드를 언제 방문하느야에 따라 아래와 같이 세 가지 방법이 있다.
# preorder  : 부모 => 왼쪽 자식 => 오른쪽 자식: 12 =>  6 =>  3 => 10 => 20 => 15 => 25
# inorder   : 왼쪽 자식 => 부모 => 오른쪽 자식:  3 =>  6 => 10 => 12 => 15 => 20 => 25: 오름차순 정렬
# postorder : 왼쪽 자식 => 오른쪽 자식 => 부모:  3 => 10 =>  6 => 15 => 25 => 20 => 12
```
